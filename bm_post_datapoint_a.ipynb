```python
import requests
from datetime import datetime

# Function to add a datapoint to a goal
def add_datapoint(username, goal_slug, access_token, value, comment, date_str=None):
    url = f"https://www.beeminder.com/api/v1/users/{username}/goals/{goal_slug}/datapoints.json"

    data = {
        'auth_token': access_token,
        'value': value,
        'comment': comment,
    }

    # Convert YYYY-MM-DD string to Unix timestamp if provided
    if date_str is not None:
        dt = datetime.strptime(date_str, "%Y-%m-%d")
        data['timestamp'] = int(dt.timestamp())

    response = requests.post(url, data=data)

    if response.status_code == 200:
        print(f"Datapoint of {value} added successfully with {date_str} as the date!")
        return response.json()
    else:
        print(f"Error: {response.status_code} - {response.json()}")
        return response.json()
```
